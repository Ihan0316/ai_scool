```python
from textwrap import indent

import cv2
import numpy as np
from matplotlib import pyplot as plt

src = cv2.imread('image/pier39.JPG', cv2.IMREAD_GRAYSCALE)
kernel =  np.ones((3,3), dtype=np.float64)/9.
dst = cv2.filter2D(src, -1, kernel)

plt.figure(figsize=(20,5))

plt.subplot(121)
plt.imshow(src, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(dst, cmap='gray')
plt.title('Averaging Filter')
plt.axis('off')

plt.show
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('image/pier39.JPG')

plt.figure(figsize=(20,5))

for i in range(1, 4):
    k = i*2+1
    dst = cv2.blur(img, (k, k))
    desc = 'k = {}'.format(k)
    cv2.putText(dst, desc, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1.0, 0, 1, cv2.LINE_AA)

    index = 140+i

    plt.subplot(index)
    plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
    plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('image/pier39.JPG')

plt.figure(figsize=(20,5))

for i in range(1, 4):
    k = i*2+1
    dst = cv2.GaussianBlur(img, (k, k), 0)
    desc = 'k = {}'.format(k)
    cv2.putText(dst, desc, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1.0, 0, 1, cv2.LINE_AA)

    index = 140+i

    plt.subplot(index)
    plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
    plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('image/pier39.JPG', cv2.IMREAD_GRAYSCALE)

blur = cv2.GaussianBlur(img, (5, 5), 0)
sharp = cv2.addWeighted(img, 1.5, blur, -0.5, 0)

plt.figure(figsize=(20,5))

plt.subplot(131)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(132)
plt.imshow(blur, cmap='gray')
plt.title('Gaussian Blur')
plt.axis('off')

plt.subplot(133)
plt.imshow(sharp, cmap='gray')
plt.title('Sharp Image')
plt.axis('off')

plt.show()
```

```python
import cv2

src = cv2.imread('image/beach.jpg', cv2.IMREAD_GRAYSCALE)
dst = cv2.bilateralFilter(src, -1, 10, 5)

plt.figure(figsize=(20,5))

plt.subplot(121)
plt.imshow(src, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(dst, cmap='gray')
plt.title('Bilateral Filter')
plt.axis('off')

plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

src = cv2.imread('image/median.jpg')
dst = cv2.medianBlur(src, 5)

plt.figure(figsize=(20,5))

plt.subplot(121)
plt.imshow(src)
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(dst)
plt.title('Median Blur')
plt.axis('off')

plt.show()
```
