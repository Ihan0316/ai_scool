```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import matplotlib.pyplot as plt
# 한글 폰트
plt.rcParams['font.family'] = 'AppleGothic'

# 데이터 불러오기
california = fetch_california_housing(as_frame=True)
X = california.data
y = california.target

# train/val/test로 분할
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, random_state=50)  # 0.25 * 0.8 = 0.2

# 학습 데이터 정규화
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# 검증/테스트 변환
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# 모델
models = [
    ("LinearRegression", LinearRegression()),
    ("SVM", SVR()),
    ("KNN", KNeighborsRegressor())
]

# 테스트 결과
for name, model in models:
    model.fit(X_train_scaled, y_train)
    y_val_pred = model.predict(X_val_scaled)
    y_test_pred = model.predict(X_test_scaled)

    train_score = model.score(X_train_scaled, y_train)
    val_score = model.score(X_val_scaled, y_val)
    test_score = model.score(X_test_scaled, y_test)

    mse = mean_squared_error(y_val, y_val_pred)
    mae = mean_absolute_error(y_val, y_val_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_val, y_val_pred)

    print(f"------{name}------")
    print(f"Train score : {train_score:.2f}")
    print(f"Validation score : {val_score:.2f}")
    print(f"Test score : {test_score:.2f}")
    print(f"[검증셋 기준] MSE: {mse:.2f}")
    print(f"[검증셋 기준] MAE: {mae:.2f}")
    print(f"[검증셋 기준] RMSE: {rmse:.2f}")
    print(f"[검증셋 기준] R2: {r2:.2f}\n")
```
