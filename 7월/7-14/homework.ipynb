```python
import random
import numpy as np
import matplotlib.pyplot as plt

# 랜덤한 128*128 이미지 1장 rgb로 생성
def random_image():
    return [random.randint(0, 255) for _ in range(128*128*3)]

plt.imshow(np.array(random_image()).reshape(128, 128, 3))
plt.axis('off')
plt.title('Random Image')
plt.show()
```
