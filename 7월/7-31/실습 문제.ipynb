```python
import cv2
import numpy as np
from jupyter_core.migrate import src
from matplotlib import pyplot as plt
from networkx.algorithms.bipartite.basic import color

src1 = cv2.imread('image/cat.jpg', cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread('image/dog.jpg', cv2.IMREAD_GRAYSCALE)
src3 = cv2.imread('image/cat.jpg', cv2.IMREAD_COLOR)
```

```python
# 문제 1.
dst1 = src1 + 100
dst2 = cv2.add(src1, 100)

plt.figure(figsize=(20, 10))

plt.subplot(231)
plt.imshow(src1, cmap='gray')
plt.title('original')
plt.axis('off')

plt.subplot(232)
plt.imshow(dst1, cmap='gray')
plt.title('numpy add')
plt.axis('off')

plt.subplot(233)
plt.imshow(dst2, cmap='gray')
plt.title('cv2 add')
plt.axis('off')
```

```python
# 문제 2.
dst3 = cv2.addWeighted(src1, 0.6, src2, 0.4, 0)

plt.figure(figsize=(20, 10))

plt.subplot(234)
plt.imshow(dst3, cmap='gray')
plt.title('weighted add')
plt.axis('off')
```

```python
# 문제 3.
# np.clip(이미지원본 + 더할 값(이미지도 가능, 이때는 numpy add로 작동), 0, 255)
dst4 = np.clip(src1 + src2, 0, 255)

plt.figure(figsize=(20, 10))

plt.subplot(235)
plt.imshow(dst4, cmap='gray')
plt.title('clip')
plt.axis('off')
```

```python
# 문제 4.
b, g, r = cv2.split(src3)

image_red = cv2.merge([r, r, r])
image_green = cv2.merge([g, g, g])
image_blue = cv2.merge([b, b, b])

plt.figure(figsize=(20, 10))

plt.subplot(231)
plt.imshow(image_red, cmap='gray')
plt.title('red')
plt.axis('off')

plt.subplot(232)
plt.imshow(image_green, cmap='gray')
plt.title('green')
plt.axis('off')

plt.subplot(233)
plt.imshow(image_blue, cmap='gray')
plt.title('blue')
plt.axis('off')
```

```python
# 문제 5.
colors = ('blue', 'green', 'red')

plt.figure(figsize=(20, 10))

plt.subplot(221)
plt.imshow(cv2.cvtColor(src3, cv2.COLOR_BGR2RGB))
plt.title('original')
plt.axis('off')

plt.subplot(222)
for i, color in enumerate(colors):
    hist = cv2.calcHist([src3], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)
    plt.xlim([0, 256])
plt.title('Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()
```

```python
# 문제 6.
dst = cv2.normalize(src1, None, 0, 255, cv2.NORM_MINMAX)
hist_src = cv2.calcHist([src1], [0], None, [256], [0, 256])
hist_dst = cv2.calcHist([dst], [0], None, [256], [0, 256])

plt.figure(figsize=(20, 10))

plt.subplot(221)
plt.imshow(cv2.cvtColor(src1, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(222)
plt.plot(hist_src, color='black')
plt.title('Histogram_org')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.grid(True)

plt.subplot(223)
plt.imshow(dst, cmap='gray')
plt.title('equalizeHist Image')
plt.axis('off')

plt.subplot(224)
plt.plot(hist_dst, color='black')
plt.title('Histogram_equalizeHist')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.grid(True)
```

```python
# 문제 7.
dst = cv2.equalizeHist(src1)
hist_src = cv2.calcHist([src1], [0], None, [256], [0, 256])
hist_dst = cv2.calcHist([dst], [0], None, [256], [0, 256])

plt.figure(figsize=(20, 10))

plt.subplot(221)
plt.imshow(cv2.cvtColor(src1, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(222)
plt.plot(hist_src, color='black')
plt.title('Histogram_org')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.grid(True)

plt.subplot(223)
plt.imshow(dst, cmap='gray')
plt.title('equalizeHist Image')
plt.axis('off')

plt.subplot(224)
plt.plot(hist_dst, color='black')
plt.title('Histogram_equalizeHist')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.grid(True)
```
