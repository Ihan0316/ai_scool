```python
import sys
import cv2
from matplotlib import pyplot as plt

img1 = cv2.imread('image/cat.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('image/cat.jpg', cv2.IMREAD_COLOR)

h, w = img2.shape[:2]
print('img1.shape : ', img1.shape)
print('img2.shape : ', img2.shape)

for y in range(300, h):
    for x in range(300, w):
        img1[y, x] = 255
        img2[y, x] = (0, 0, 255)

img1[300:640, 300:640] = 255
img2[300:640, 300:640] = (0, 0, 255)

plt.figure(figsize=(12, 5))

plt.subplot(121), plt.axis('off'), plt.imshow(img1, cmap='gray'), plt.title('img1')
plt.subplot(122), plt.axis('off'), plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)), plt.title('img2')

plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

src1 = cv2.imread('image/cat.jpg', cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread('image/dog.jpg', cv2.IMREAD_GRAYSCALE)

dst1 = cv2.add(src1, src2)
dst2 = cv2.addWeighted(src1, 0.0, src2, 0.2, 0.0)
dst3 = cv2.addWeighted(src1, 0.5, src2, 0.5, 0.0)
dst4 = cv2.addWeighted(src1, 0.2, src2, 0.8, 0.0)

plt.figure(figsize=(12, 5))

plt.subplot(231), plt.axis('off'), plt.imshow(src1, cmap='gray'), plt.title('cat')
plt.subplot(232), plt.axis('off'), plt.imshow(src2, cmap='gray'), plt.title('dog')
plt.subplot(233), plt.axis('off'), plt.imshow(dst1, cmap='gray'), plt.title('add')
plt.subplot(234), plt.axis('off'), plt.imshow(dst2, cmap='gray'), plt.title('0.0_0.2')
plt.subplot(235), plt.axis('off'), plt.imshow(dst3, cmap='gray'), plt.title('0.5_0.5')
plt.subplot(236), plt.axis('off'), plt.imshow(dst4, cmap='gray'), plt.title('0.2_0.8')

plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

src1 = cv2.imread('image/abs1.jpg', cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread('image/abs2.jpg', cv2.IMREAD_GRAYSCALE)

dst = cv2.absdiff(src1, src2)

plt.figure(figsize=(12, 5))

plt.subplot(131), plt.axis('off'), plt.imshow(src1, 'gray'), plt.title('abs1')
plt.subplot(132), plt.axis('off'), plt.imshow(src2, 'gray'), plt.title('abs2')
plt.subplot(133), plt.axis('off'), plt.imshow(dst, 'gray'), plt.title('result')

plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread('image/candies.png')

b, g, r = cv2.split(image)

merge_b = cv2.merge([b, b, b])
merge_g = cv2.merge([g, g, g])
merge_r = cv2.merge([r, r, r])

plt.figure(figsize=(12, 5))

plt.subplot(221), plt.axis('off'), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('original')
plt.subplot(222), plt.axis('off'), plt.imshow(merge_b, cmap='gray'), plt.title('b')
plt.subplot(223), plt.axis('off'), plt.imshow(merge_g, cmap='gray'), plt.title('g')
plt.subplot(224), plt.axis('off'), plt.imshow(merge_r, cmap='gray'), plt.title('r')

plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread('image/candies.png')
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv_image)

merge_h = cv2.merge([h, h, h])
merge_s = cv2.merge([s, s, s])
merge_v = cv2.merge([v, v, v])

plt.figure(figsize=(12, 5))

plt.subplot(221), plt.axis('off'), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('original')
plt.subplot(222), plt.axis('off'), plt.imshow(merge_h, cmap='gray'), plt.title('h')
plt.subplot(223), plt.axis('off'), plt.imshow(merge_s, cmap='gray'), plt.title('s')
plt.subplot(224), plt.axis('off'), plt.imshow(merge_v, cmap='gray'), plt.title('v')

plt.show()
```
